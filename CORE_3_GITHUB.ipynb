```python
#importing libraries.
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
pd.set_option('display.max_columns',100)
```

Loading Data.

```python
sales='/content/drive/MyDrive/CodingDojo/01-Fundamentals/Week02/sales_predictions_2023.csv'
df_sales=pd.read_csv(sales)
df_sales.head()
```

```python
#histogram to view distribution of out_establishment_year and item_outlet_sales.
ax = df_sales['Outlet_Establishment_Year'].hist(bins = 30, edgecolor = 'black')
ax.tick_params(axis='x', rotation = 45)
ax.ticklabel_format(style='plain')
ax.set_title('Distribution of establishment year and outlet sales')
ax.set_xlabel('year(s)')
ax.set_ylabel('sales($)');
```

#**intrepretation**.
1985 had the highest sales,1990 and 1995 had no sales then the sales plateaued till 2010 but with a decline before 2000.

```python
#boxplot of Item_Visibility and Item_Fat_Content.
ax = sns.boxplot(data=df_sales, x='Item_Fat_Content', y='Item_Visibility')
ax.set_title("Comparing Item_Visibility  with Item_Fat_Content");
```

#**intrepretation**
#LowFat and regular fat almost paint the same distribution with the most outlayers compared to lowfat,LF and reg

```python
#countplot of outlet establishment year.

ax = sns.countplot(data=df_sales, x='Outlet_Establishment_Year')
ax.set_title("relationship between establishment year and sales")
```

#**intrepretation**
1985 has the highest sales and 1998 has the lowest and the rest till 2009 are plateaued .

```python
#heatmap showing relationship between outlet location type and item visibility.

corr = df_sales.corr(numeric_only = True)
corr.round(2)

sns.heatmap(corr, cmap = 'Greens', annot = True);
```

#**interpretation**.
##There is a moderate correlation between item MRP and item outlet sales.
